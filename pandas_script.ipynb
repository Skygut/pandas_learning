```python
import pandas as pd
import numpy as np

# Створюємо DataFrame з 10 рядками
data = {"A": range(10), "B": np.random.randn(10), "C": np.random.randn(10)}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Ділимо DataFrame на дві частини
n = len(df) // 2  # Знаходимо середину DataFrame
df_part1 = df.iloc[:n]  # Перші n рядків
df_part2 = df.iloc[n:]  # Всі рядки від n до кінця

print("\nFirst part of DataFrame (df_part1):")
print(df_part1)

print("\nSecond part of DataFrame (df_part2):")
print(df_part2)
```

```python
import pandas as pd
import numpy as np

# Створюємо дати для індексу
dates = pd.date_range("20230101", periods=6)

# Створюємо DataFrame
data = np.random.randn(6, 4)
df10 = pd.DataFrame(data, index=dates, columns=list("ABCD"))

print("Original DataFrame:")
print(df10)

# Використовуємо loc для доступу до першого рядка
first_row = df10.loc[dates[1]]

print("\nFirst row of DataFrame:")
print(first_row)
```
